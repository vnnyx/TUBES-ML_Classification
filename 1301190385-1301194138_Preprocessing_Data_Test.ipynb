```python
# Proses import library seperti numpy, matplotlb, pandas,
# dan terdapat juga import library sklearn utntuk pre processing data

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
from sklearn import preprocessing
```

```python
# Proses read data dari dataset kendaraan_test yang disimpan dalam variabel df lalu ditampilkan data 5 teratas.

df = pd.read_csv('Dataset/kendaraan_test.csv')
df.head()
```

```python
df.describe()
```

```python
# Pembuatan variabel baru untuk menampung proses labele encoding yang pada kasus ini untuk melabeli data yang diinginkan yaitu
# Jenis kelamin, Umur Kendaraan, dan Kendaraan Rusak dari kategorikal data menjadi data numerik.

le = preprocessing.LabelEncoder()
def encode(type, change):
  df[type] = le.fit_transform(df[type])
  df[type] = df[type].replace(change, np.NaN)

encode('Jenis_Kelamin', 2)
encode('Umur_Kendaraan', 3)
encode('Kendaraan_Rusak', 2)
df.head()
```

```python
# Proses Penormalisasian menggunakan MinMaxScalar dari library sklearn.preprocessing. 

names = df.columns
scaler = preprocessing.MinMaxScaler()
d = scaler.fit_transform(df)
df = pd.DataFrame(d, columns=names)
df.head()
```

```python
# Proses Export Dataset setelah proses preprocessing.

df.to_csv('Dataset/kendaraan_test_clean.csv', index=False)
```
